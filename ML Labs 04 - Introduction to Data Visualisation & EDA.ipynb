## CODE:

```python
# Importing the libraries.
import pandas as pd
import numpy as np
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

<hr>

## Part: A)

```python
# Creating the Dataset:
X, y = make_classification(n_samples=10000, 
                           n_features=12,
                           n_informative=4,
                           n_redundant=8,
                           n_clusters_per_class=2,
                           n_classes=3,
                           random_state=123)
```

```python
df = pd.DataFrame(X, columns = ['F1','F2','F3','F4','F5','F6','F7','F8','F9','F10','F11','F12'])
```

#### Dataset Exploration

```python
df.head()
```

```python
type(df)
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

Creating the required classes in first four feature F1,F2,F3,F4.

```python
# Trying with np.where
df['F1']=np.where(df['F1']>0,"A","B")
```

```python
# Trying with creating functions
def f2_change(x):
    if(x<0):
        return("Type 1")
    else:
        return("Type 2")
```

```python
def f3_change(x):
    if(x<0):
        return("A")
    elif(x<=0.5):
        return("B")
    else:
        return("C")
```

```python
def f4_change(x):
    if(x>2):
        return("High")
    elif(x>(-0.7)):
        return("Medium")
    else:
        return("Low")
```

```python
# Applying functions
df["F2"]=df["F2"].apply(f2_change)
df["F3"]=df["F3"].apply(f3_change)
df.F4=df.F4.apply(f4_change)
```

```python
print("Value Counts:")
print("----------------------")
print("F1:")
print(df.F1.value_counts())
print("----------------------")
print("F2:")
print(df.F2.value_counts())
print("----------------------")
print("F3:")
print(df.F3.value_counts())
print("----------------------")
print("F4:")
print(df.F4.value_counts())
print("----------------------")
```

## Part B
`Exploring different kind of plots and findingout hidden patterns in dataset.`

#### Distplot
- distplot() 
- The distplot represents the univariate distribution of data i.e. data distribution of a variable against the density distribution. It accepts the data variable as an argument and returns the plot with the density distribution.

```python
sns.distplot(df["F5"])
plt.title("Dist Plot")
plt.show()
```

<hr>

#### Jointplot
- A Jointplot comprises three plots.
  - One plot displays a bivariate graph which shows how the dependent variable(Y) varies with the independent variable(X).
  - Another plot is placed horizontally at the top of the bivariate graph and it shows the distribution of the independent variable(X)

```python
sns.jointplot(x='F6', y='F7',data=df,kind='hex')
plt.title("Joint Plot")
plt.show()
```

<hr>

#### Pairplot
- pairplot()
- To plot multiple pairwise bivariate distributions in a dataset.

```python
sns.pairplot(df)
plt.title("Pair Plot")
plt.show()
```

<hr>

#### Rugplot
- A rug plot is a simple plot to explore distributions of a continous variable.

```python
plt.figure(figsize=(15,5))

plt.subplot(1,2,1)
sns.rugplot(data=df, x="F8")

plt.subplot(1,2,2)
sns.kdeplot(data = df, x = "F8")
sns.rugplot(data=df, x="F8")
plt.show()
```

<hr>

#### Catplot
- catplot shows frequencies (or optionally fractions or percents) of the categories of one, two or three categorical variables.

```python
sns.catplot(x="F3", data=df, kind='count')
plt.title("Cat Plot")
plt.show()
```

```python
cat=['High', 'Medium', 'Low']
sns.catplot(x='F4', y='F10', data=df, kind='bar',order=cat)
plt.title("Cat Plot")
plt.show()
```

```python
sns.catplot(x='F4', y='F10', data=df, kind='bar')
plt.title("Cat Plot")
plt.show()
```

<hr>

#### Barplot
-  It shows the relationship between a numeric and a categoric variable. Each entity of the categoric variable is represented as a bar. The size of the bar represents its numeric value.

```python
df.F1.value_counts().plot(kind='bar', color='brown')
plt.title("Bar Plot")
plt.show()
```

```python
df.groupby('F4').max().plot(kind='bar')
plt.title("Bar Plot")
plt.show()
```

```python
df.F3.value_counts().plot(kind='barh', color='yellow')
plt.title("Bar Plot")
plt.show()
```

<hr>

#### Countplot
- A countplot is kind of likea histogram or a bar graph for some categorical area. It simply shows the number of occurrences of an item based on a certain type of category.

```python
sns.countplot(df['F3'])
plt.title("Count Plot")
plt.show()
```

```python
sns.countplot(x='F3',data=df,hue='F2')
plt.title("Count Plot")
plt.show()
```

<hr>

#### Violinplot
- Violin plots are used when you want to observe the distribution of numeric data, and are especially useful when you want to make a comparison of distributions between multiple groups. The peaks, valleys, and tails of each group's density curve can be compared to see where groups are similar or different.

```python
sns.violinplot(df['F7'])
plt.title("Violinplot")
plt.show()
```

<hr>

#### Stripplot
- A strip plot is a single-axis scatter plot that is used to visualise the distribution of many individual one-dimensional values.

```python
sns.stripplot(x='F1',y='F5',data=df)
plt.title("Stripplot")
plt.show()     #Stripplot with jitter.
```

```python
sns.stripplot(x='F1',y='F5',data=df,jitter=False)
plt.title("Stripplot")
plt.show()      #Stripplot without jitter.
```

<hr>

#### Swarmplot
- A swarm plot is another way of plotting the distribution of an attribute or the joint distribution of a couple of attributes.

```python
sns.swarmplot(x='F1', y='F6', data=df)
plt.title("Swarmplot")
plt.show()  # no over lapping.
```

<hr>

#### FacetGrid 
- FacetGrid object takes a dataframe as input and the names of the variables that will form the row, column, or hue dimensions of the grid. The variables should be categorical and the data at each level of the variable will be used for a facet along that axis.

```python
# FacetGrid of histplot of a feature
f=sns.FacetGrid(df, col='F2')
f.map(plt.hist, "F5")
plt.show()
```

```python
# FacetGrid of scatterplot of a feature
f=sns.FacetGrid(df, col='F2',hue='F4')
f.map(plt.scatter, "F5","F7")
plt.show()
```

<hr>

#### Lmplot
- Seaborn's lmplot is a 2D scatterplot with an optional overlaid regression line. This is useful for comparing numeric variables.

```python
sns.lmplot(x='F6', y='F12', data=df,hue='F1',fit_reg=False)
plt.title("Lm Plot")
plt.show()
```

```python
# Loading Iris dataset.
from sklearn.datasets import load_iris
iris= load_iris()
```

#### Scatterplot
- A scatter plot is a type of plot or mathematical diagram using Cartesian coordinates to display values for typically two variables for a set of data.

```python
sns.scatterplot(x='sepal length (cm)', y='sepal width (cm)', data=iris)
plt.title("Scatter Plot")
plt.show()
```

```python
sns.scatterplot(x='sepal length (cm)', y='sepal width (cm)', data=iris, hue='Species')
plt.title("Scatter Plot")
plt.show()
```

#### Scatter 3D

```python
z = df.F6
x = df.F7
y = df.F9

# Creating figure
fig = plt.figure(figsize = (10, 7))
ax = plt.axes(projection ="3d")
 
# Creating plot
ax.scatter3D(x, y, z, color = "green")
plt.title("simple 3D scatter plot")
 
# show plot
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')

x = iris['sepal length (cm)']
y = iris['sepal width (cm)']
z = iris['petal length (cm)']

ax.set_xlabel("Sepal length")
ax.set_ylabel("Sepal width")
ax.set_zlabel("Petal length")

ax.scatter(x, y, z)
plt.show()
```

<hr>

#### Heatmap
- A heatmap is a graphical representation of data that uses a system of color-coding to represent different values. Heatmaps are used in various forms of analytics but are most commonly used to show user behavior on specific features in dataset.

```python
# Finding corelation between the features.
iris_corr=iris.corr()
```

```python
# Heatmap:
sns.heatmap(iris_corr,annot=True, cmap='ocean')
plt.title("Heatmap")
plt.show()
```

<hr>

#### Boxplot
- A boxplot is a standardized way of displaying the distribution of data based on a five number summary (“minimum”, first quartile (Q1), median, third quartile (Q3), and “maximum”). It can tell you about your outliers and what their values are.

```python
# Boxplot of one feature.
sns.boxplot(iris['sepal length (cm)'])
```

```python
# Boxplot of two features consisting of 3 classes.
sns.boxplot(x='Species', y='sepal length (cm)',data=iris)
plt.title("Boxplot of different Species")
plt.show()
```

<hr>

#### Kdeplot
- Kdeplot is a Kernel Distribution Estimation Plot which depicts the probability density function of the continuous or non-parametric data variables.

```python
# Drawing kdeplot on one continuous plot.
sns.kdeplot(iris['sepal length (cm)'])
plt.title("Kdeplot")
plt.show()
```
